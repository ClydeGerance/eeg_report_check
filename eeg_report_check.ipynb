```python
# Initialize an empty list to collect all data_person DataFrames
all_data_person = []

# Example filename_metric
filename_metric = "Data Set EEG ONLY v2 (Day 1 - Engagement) - Sheet1.csv"

# Extract day number from filename_metric
if 'Day 1' in filename_metric:
    day_number = 1
elif 'Day 2' in filename_metric:
    day_number = 2
elif 'Day 3' in filename_metric:
    day_number = 3
else:
    raise ValueError("Day number not found in filename.")

# Define the regex pattern to match the day number in filename_person
day_pattern = f'D{day_number}'

# Iterate through each file in root_dir_person
for filename_person in os.listdir(root_dir_person):
    # Check if the file matches the day pattern
    if day_pattern in filename_person:
        # Step 1: Extract the metric name from the filename using a regex
        metric_name_match = re.search(r'Day \d+ - (\w+)', filename_metric)
        if metric_name_match:
            metric_name = metric_name_match.group(1)
        else:
            raise ValueError("Metric name not found in the filename")

        # Define the columns to keep based on the extracted metric name
        columns_to_keep = ["Timestamp", f"PM.{metric_name}.Min", f"PM.{metric_name}.Max"]

        # Step 2: Read the CSV and keep only the relevant columns
        data_person = pd.read_csv(os.path.join(root_dir_person, filename_person), header=1)

        # Keep only the relevant columns
        data_person = data_person[columns_to_keep]

        # Drop rows with null values in any of the selected columns
        data_person = data_person.dropna()

        # Apply floor to Timestamp column (if necessary)
        data_person["Timestamp"] = data_person["Timestamp"].astype(float)
        data_person["Timestamp"] = data_person["Timestamp"].apply(lambda x: f"{x:.0f}")

        # Append the processed DataFrame to the list
        all_data_person.append(data_person)

# Concatenate all data_person DataFrames into one DataFrame
combined_data_person = pd.concat(all_data_person, ignore_index=True)

# Display the combined DataFrame
combined_data_person
```

```python
# Initialize an empty list to store the 2D array
array_person = []

# Iterate through each row of the DataFrame
for index, row in combined_data_person.iterrows():
    # Append [Timestamp, PM.Attention.Min] to the 2D array
    array_person.append([row['Timestamp'], row[f"PM.{metric_name}.Min"]])
    # Append [Timestamp, PM.Attention.Max] to the 2D array
    array_person.append([row['Timestamp'], row[f"PM.{metric_name}.Max"]])

# Convert the list of lists to a NumPy array
array_person = np.array(array_person)

# Convert array_person to a DataFrame
df_person = pd.DataFrame(array_person, columns=['Timestamp', 'Value'])

# Drop duplicates based on Timestamp and PM.Attention.Value
df_person = df_person.drop_duplicates()

df_person = df_person.sort_values(by='Value')
# Sort by Timestamp in ascending order
df_person = df_person.sort_values(by='Timestamp')

# Convert back to NumPy array if needed
array_person = df_person.values

# Print the 2D array with Timestamp sorted in ascending order
print(array_person)
```

```python
def find_non_matching_indices(array_person, array_metric):
    person_matched = True
    metric_matched = True

    for person_item in array_person:
        person_string = person_item[0]
        person_value = round(float(person_item[1]), 1)  # Convert to float and round to nearest tenth
        found_match = False
        
        for metric_item in array_metric:
            metric_string = metric_item[0]
            metric_value = round(metric_item[1], 1)  # Round to nearest tenth
            
            if person_string == metric_string and person_value == metric_value:
                found_match = True
                break
        
        if not found_match:
            person_matched = False
            print(f"No matching value found for person item '{person_item}'.")

    for metric_item in array_metric:
        metric_string = metric_item[0]
        metric_value = round(metric_item[1], 1)  # Round to nearest tenth
        found_match = False
        
        for person_item in array_person:
            person_string = person_item[0]
            person_value = round(float(person_item[1]), 1)  # Convert to float and round to nearest tenth
            
            if metric_string == person_string and metric_value == person_value:
                found_match = True
                break
        
        if not found_match:
            metric_matched = False
            print(f"No matching value found for metric item '{metric_item}'.")

    if person_matched and metric_matched:
        print("Everything is matched.")

find_non_matching_indices(array_person, array_metric)
```
